# Cost Sensitive

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics  
from sklearn.svm import SVC
from sklearn.tree import export_graphviz
from IPython.display import Image  
import pydotplus
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/Titanic_final.csv", index_col=0)
X = df.drop('Survived', axis=1)
y = df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) 
```

Confusion matrix: 

|  | predicted negative | predicted positive |
|:--- | :---: | :---: |
| actual negative | TN | FP|
| actual positive | FN | TP|

### $ Loss = fpr\cdot cfp + fnr\cdot cfn $ <br>
- $cfp$: false positive cost
- $cfn$: false negative cost
- $fpr = \frac{fp}{fp+tn}$
- $fnr = \frac{fn}{fn+tp}$

```python
def loss_calc(cm, cfp, cfn):
    fpr = cm[0][1]/(cm[0][1]+cm[0][0])
    fnr = cm[1][0]/(cm[1][0]+cm[1][1])
    return fpr*cfp + fnr*cfn
```

Define misclasification cost:

```python
cfp = 1
cfn = 10
```

## Cost Sensitive using class_weight
class_weight is a parameter in some of sklearn classifiers. It penalizes mistakes in samples of class[i] with class_weight[i] instead of 1. Therefore higher class-weight means you want to put more emphasis on a class.

#### Decision Tree Classifier

```python
clf = DecisionTreeClassifier(criterion='entropy',max_depth=4, random_state=42)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
cm = metrics.confusion_matrix(y_test,y_pred)
acc = metrics.accuracy_score(y_test,y_pred)
print('Accuracy: %.2f' %acc)
pd.DataFrame(cm)
```

#### Cost Sensitive Desicion Tree Classifier

```python
clf_w = DecisionTreeClassifier(criterion='entropy',max_depth=4, class_weight={0:cfp,1:cfn}, random_state=42)
clf_w.fit(X_train,y_train)
y_pred_w = clf_w.predict(X_test)
cm_w = metrics.confusion_matrix(y_test,y_pred_w)
acc_w = metrics.accuracy_score(y_test,y_pred_w)
print('Accuracy: %.2f' %acc_w)
pd.DataFrame(cm_w)
```

As we can see, the regular DT has higher accuracy, but is it better in terms of cost?

```python
loss = loss_calc(cm, cfp, cfn)
loss_w = loss_calc(cm_w, cfp, cfn)
if loss > loss_w:
    print('The Cost sensitive DT loss is better')
else:
    print('The regular DT loss is better')
print('Decision Tree Classifier loss: %.2f' %loss)
print('Cost Sensitive Desicion Tree Classifier loss: %.2f' %loss_w)
```

## Cost Sensitive using threshold
$ t = \frac{cfp}{cfp+cfn}$

```python
t = cfp/(cfp+cfn)
t
```

```python
y_prob = clf.predict_proba(X_test)
y_pred_t = [0 if p[1]<t else 1 for p in y_prob]
cm_t = metrics.confusion_matrix(y_test,y_pred_t)
acc_t = metrics.accuracy_score(y_test,y_pred_t)
loss_t = loss_calc(cm_t,cfp,cfn)
pd.DataFrame(cm_t)
```

```python
print('Accuracy:')
print('Decision Tree: %.2f' % (acc*100))
print('class_weight method: %.2f' % (acc_w*100)) 
print('threshold method: %.2f' % (acc_t*100) )
print('------------------------------------')
print('Loss:')
print('Decision Tree: %.2f' % loss)
print('class_weight method: %.2f' % loss_w) 
print('threshold method: %.2f' % loss_t)
```

#### The impact of changing the threshold on accuracy and loss

```python
acc_list = []
loss_list = []
cm_list = []
for t in np.arange(0, 1, 0.01):
    y_pred_tl = [0 if p[1]<t else 1 for p in y_prob]
    cm_tl = metrics.confusion_matrix(y_test,y_pred_tl)
    loss_tl = loss_calc(cm_tl,cfp,cfn)
    loss_list.append(loss_tl)
    acc_tl = metrics.accuracy_score(y_test,y_pred_tl)
    acc_list.append(acc_tl)
```

```python
fig,ax = plt.subplots()
ax.plot(np.arange(0, 1, 0.01), loss_list, color="red")
ax.set_xlabel("Threshold",fontsize=14)
ax.set_ylabel("Loss",color="red",fontsize=14)
ax2=ax.twinx()
ax2.plot(np.arange(0, 1, 0.01),acc_list,color="blue")
ax2.set_ylabel("Accuracy",color="blue",fontsize=14)
plt.show()
```

## Cost sensitive for imalanced data

The misclassification cost can represent the imbalance of the data set.
- $cfp = \frac{N}{N+P}$
- $cfn = \frac{P}{P+N}$

Changing the class distributions **only** to illustrate the imbalance problem.

```python
sampled_data = pd.concat([df[df['Survived']==0].sample(900,replace=True,random_state = 42), 
                          df[df['Survived']==1].sample(100, random_state = 42)])
X1 = sampled_data.drop('Survived', axis=1)
y1 = sampled_data['Survived']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.25, stratify=y1, random_state=42) 
```

```python
N = len(y1[y1 == 0])
P = len(y1[y1 == 1])
cfp1 = N/(N+P)
cfn1 = P/(P+N)
print(cfp1, cfn1)
print(N,P)
```

```python
clf_w1 = DecisionTreeClassifier(criterion='entropy',max_depth=4,class_weight={0:cfn1,1:cfp1},random_state=42)
clf_w1.fit(X_train1,y_train1)
y_pred1 = clf_w1.predict(X_test1)
acc1 = metrics.accuracy_score(y_test1,y_pred1)
print(acc1)
cm1 = metrics.confusion_matrix(y_test1,y_pred1)
pd.DataFrame(cm1)
```

```python
print(metrics.classification_report(y_test1,y_pred1))
```

In that case, loss corresponds to error rate, while accuracy equal to $1-loss$.

```python
1- loss_calc(cm1, cfp1, cfn1)
```

When using class_weight='balanced' the weights are calculate to get the data balanced.

```python
clf_w2 = DecisionTreeClassifier(criterion='entropy',max_depth=4,class_weight='balanced',random_state=42)
clf_w2.fit(X_train1,y_train1)
y_pred2 = clf_w2.predict(X_test1)
acc2 = metrics.accuracy_score(y_test1,y_pred2)
print(acc2)
cm2 = metrics.confusion_matrix(y_test1,y_pred2)
pd.DataFrame(cm2)
```

Using threshold for imbalanced

```python
t1 = cfn1/(cfp1+cfn1)
clf.fit(X_train1,y_train1)
y_prob1 = clf.predict_proba(X_test1)
y_pred_t1 = [0 if p1[1]<t1 else 1 for p1 in y_prob1]
cm_t1 = metrics.confusion_matrix(y_test1,y_pred_t1)
acc_t1 = metrics.accuracy_score(y_test1,y_pred_t1)
loss_t1 = loss_calc(cm_t1,cfp1,cfn1)
print(acc_t1)
pd.DataFrame(cm_t1)
```
